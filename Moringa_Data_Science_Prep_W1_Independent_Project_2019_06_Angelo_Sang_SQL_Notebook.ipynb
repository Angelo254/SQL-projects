```python
#Display a list of Uganda_Karamoja_Subcounty_Crop_Yield_Population
%load_ext sql
%sql sqlite://
```

```python
# Importing the python libraries that we will need in this notebook
#
import csv
import pandas as pd
```

```python
# Loading the table from a local CSV file
#
with open('Uganda_Karamoja_Subcounty_Crop_Yield_Population.csv','r') as f:
    Uganda_Karamoja_Subcounty_crop_yield_population = pd.read_csv(f, index_col=0, encoding='utf-8') 

# Then storing it in an SQL table of our in memory sqlite database
#
%sql DROP TABLE if EXISTS Uganda_Karamoja_Subcounty_crop_yield_population;
%sql PERSIST Uganda_Karamoja_Subcounty_crop_yield_population;
```

```python
#Display a list of Sub Counties and their population and areas.
%%sql
SELECT 'SUBCOUNTY', "POP","Area" FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
```

```python
#Sort the list of districts by total crop area (descending order)
%%sql
SELECT "DISTRICT_NAME", "Crop_Area_Ha" FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
ORDER BY "Crop_Area_Ha" DESC;
```

```python
#Select only the Sub counties from the Moroto district, order them alphabetically 
#and show their production of sorghum
%%sql
SELECT "SUBCOUNTY_NAME", "DISTRICT_NAME", "S_Prod_Tot" FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
WHERE DISTRICT_NAME = "MOROTO"
ORDER BY "DISTRICT_NAME" ASC
```

```python
#Compute the total Maize production per District
%%sql
SELECT "DISTRICT_NAME", "M_Prod_Tot" FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
```

```python
#Compute the number of Sub counties where Maize is produced 
#and the total Maize production per District
%%sql
SELECT "SUBCOUNTY_NAME", "DISTRICT_NAME", "M_Area_Ha", "M_Prod_Tot" FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
WHERE "M_Area_Ha" > 0
```

```python
#Compute the overall Crop area in all Sub counties where population is over 20000
%%sql
SELECT "SUBCOUNTY_NAME", "Crop_Area_Ha" FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
WHERE "POP" > 20000
```

```python
#Sort the Maize production in descending order by Districts,
#only taking into account Sub counties where Maize area is larger than Sorghum area,
#and display the number of Sub counties per district matching that criteria
%%sql
SELECT "SUBCOUNTY_NAME", "DISTRICT_NAME", "S_Area_Ha", "M_Area_Ha" FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
WHERE "S_Area_Ha" < "M_Area_Ha"
ORDER BY "DISTRICT_NAME" DESC
```
