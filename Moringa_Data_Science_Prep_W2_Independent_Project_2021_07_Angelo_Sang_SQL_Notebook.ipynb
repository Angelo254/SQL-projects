```python
# We will first load the sql extension into our environment
# Connecting to our Database
%load_ext sql
 
%sql sqlite://
```

```python
# Loading our Datasets
# Importing Pandas
import pandas as pd
```

```python
# Loading our lighting dataset  from the respective CSV file 
with open('GrandElectors_by_state.csv','r') as f:
    grandelectors = pd.read_csv(f, index_col=0, encoding='utf-8')
%sql DROP TABLE if EXISTS grandelectors;
%sql PERSIST grandelectors;
%sql SELECT * FROM grandelectors LIMIT 5;
```

```python
# Loading our lighting dataset  from the respective CSV file 
with open('Population_by_state.csv','r') as f:
    population = pd.read_csv(f, index_col=0, encoding='utf-8')
%sql DROP TABLE if EXISTS population;
%sql PERSIST population;
%sql SELECT * FROM population LIMIT 5;
```

```python
# Converting the states in GrandElectors_by_state table to uppercase
%%sql
update grandelectors
set 
State = upper(State)
```

```python
#Joining the tables on the state key
%%sql
SELECT grandelectors.State,Population,GrandElectors 
FROM grandelectors INNER JOIN Population 
ON grandelectors = Population.State LIMIT 5;
```

```python
# change the name of the "District of Columbia" state to its short version "DC"

%%sql
UPDATE grandelectors
SET State = "DC" 
WHERE State = "DISTRICT OF COLUMBIA";
SELECT * FROM grandelectors;
```

```python
#compute the ratio between the number of grand electors and the population.
%%sql
ALter table grandelectors
add Ratio varchar  (255);
Select grandelectors, (Count(GrandElectors)* 100 / (Select Count(Population) From population)) as Ratio
From grandelectors
Group By grandelectors
```

```python
#order the states by decreasing ratio of Grand Electors per capita.
%%sql
SELECT State, Ratio
FROM grandelectors
order by  Ratio desc;
```

```python
#To compute the running total of Grand Electors in that sorted list.
%%sql
select GrandElectors, State
    GrandElectors,
    (SELECT Sum(GrandElectors)/2 FROM grandelectors)
FROM grandelectors
group by GrandElectors
```

```python
#Independently, to compute the half of the total of Grand Electors overall (in the whole country):
%%sql
select avg(GrandElectors)
from grandelectors;
```

```python
#To filter our sorted list of states in order to keep only the (top) ones enabling us to reach the computed threshold. (the other states can be ignored)
%%sql
SELECT State,GrandElectors
from grandelectors 
where GrandElectors > 10 ;
```
